## popularity based recommendation system

```python
#merging ratings and books data set
ratings_with_name=ratings.merge(books,on='ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()##no of rating
num_rating_df.rename(columns={'Book-Rating':'num-rating'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].agg(lambda x: x.astype(float).mean()).reset_index()
avg_rating_df.rename(columns = {'Book-Rating' : 'avg_rating'}, inplace = True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num-rating']>=250]
popular_df
```

```python
popular_df=popular_df.sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','Image-URL-M','num-rating','avg_rating']]
```

```python
popular_df
```

```python
popular_df['Image-URL-M'] = popular_df['Image-URL-M'].replace('http', 'https', regex=True)
```

```python
popular_df
```

## Collabarative Filtering Based Recommender System

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
user_with_exp = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(user_with_exp)]
```

```python
books_with_minratings = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >50
```

```python
books_with_minratings=books_with_minratings[books_with_minratings].index
```

```python
books_with_minratings
```

```python
filtered_rating=filtered_rating[filtered_rating['Book-Title'].isin(books_with_minratings)]
```

```python
filtered_rating
```

```python
pt=filtered_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace = True)
```
